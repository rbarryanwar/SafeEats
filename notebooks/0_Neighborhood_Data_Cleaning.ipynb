```python
import xlrd
all_data = pd.DataFrame()
for f, g in zip(glob.glob('*_NeighborhoodDataProfile.xlsx'), sheets):
    df = pd.read_excel(f, sheet_name = g)
    all_data = all_data.append(df, ignore_index=True)
```

```python
Hoods = all_data[['Community District',  'Name','Indicator Category','Indicator',' Indicator Description',2017, '2017/18 Rank']]
Hoods = Hoods.sort_values('Community District')
Hoods.head()
Hoods.tail()
```

```python
Hoods_use.drop('2017/18 Rank', axis=1, inplace = True)
Hoods_use.drop(' Indicator Description', axis=1, inplace = True)
Hoods_use.drop('Indicator Category', axis=1, inplace = True)
Hoods_use['District Name'] = Hoods_use['Community District'] + ' ' + Hoods_use['Name']
Hoods_use.head()
```
